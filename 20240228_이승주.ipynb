```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv("/content/sample_data/train.csv")
test = pd.read_csv("/content/sample_data/test.csv")
```

```python
train
```

```python
test
```

```python
display(train.isnull().sum())
display(test.isnull().sum())
display(train.isnull().sum() / train.shape[0] * 100)
display(test.isnull().sum() / test.shape[0] * 100)
```

```python
display(train.dtypes)
display(test.dtypes)
```

```python
train[train['hour_bef_temperature'].isnull()]
```

```python
train.drop([934, 1035], axis=0, inplace=True)
train[train['hour_bef_temperature'].isnull()]
```

```python
test[test['hour_bef_temperature'].isnull()]
```

```python
test.drop([653], axis=0, inplace=True)
test[test['hour_bef_temperature'].isnull()]
```

```python
train['hour_bef_windspeed'].fillna(train['hour_bef_windspeed'].median(), inplace=True)
train['hour_bef_ozone'].fillna(train['hour_bef_ozone'].median(), inplace=True)
train['hour_bef_pm10'].fillna(train['hour_bef_pm10'].median(), inplace=True)
train['hour_bef_pm2.5'].fillna(train['hour_bef_pm2.5'].median(), inplace=True)

test['hour_bef_ozone'].fillna(test['hour_bef_ozone'].median(), inplace=True)
test['hour_bef_pm10'].fillna(test['hour_bef_pm10'].median(), inplace=True)
test['hour_bef_pm2.5'].fillna(test['hour_bef_pm2.5'].median(), inplace=True)
```

```python
display(train.isnull().sum())
display(test.isnull().sum())
```

```python
train.groupby('hour').mean()['hour_bef_windspeed']
```

```python
train.groupby('hour').mean()['hour_bef_windspeed'].plot()
```

```python
for i in train[train['hour_bef_windspeed'].isnull()].index:
  train.loc[i, 'hour_bef_windspeed'] = train.groupby('hour').mean()['hour_bef_windspeed'][train.loc[i]['hour']]
```

```python
display(train.groupby('hour').mean()['hour_bef_ozone'])
display(test.groupby('hour').mean()['hour_bef_ozone'])
```

```python
plt.plot(train.groupby('hour').mean()['hour_bef_ozone'], 'o-')
plt.plot(test.groupby('hour').mean()['hour_bef_ozone'], 'o-')
plt.show()
```

```python
for i in train[train['hour_bef_ozone'].isnull()].index:
  train.loc[i, 'hour_bef_ozone'] = train.groupby('hour').mean()['hour_bef_ozone'][train.loc[i, 'hour']]
  if train.loc[i, 'hour'] == 1:
    train.loc[i, 'hour_bef_ozone'] = (train.groupby('hour').mean()['hour_bef_ozone'][0]+ train.groupby('hour').mean()['hour_bef_ozone'][2]) / 2
for i in test[test['hour_bef_ozone'].isnull()].index:
  test.loc[i, 'hour_bef_ozone'] = test.groupby('hour').mean()['hour_bef_ozone'][test.loc[i, 'hour']]
  if test.loc[i, 'hour'] == 1:
    test.loc[i, 'hour_bef_ozone'] = (test.groupby('hour').mean()['hour_bef_ozone'][0]+ test.groupby('hour').mean()['hour_bef_ozone'][2]) / 2
```
