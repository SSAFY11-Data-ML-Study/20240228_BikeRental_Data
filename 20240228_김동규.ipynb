
### 데이터 로드


```python
data_folder = '/content/gdrive/MyDrive/data/BikeRental_Data/sample_data/'
train = pd.read_csv(data_folder+'train2.csv')
test = pd.read_csv(data_folder+'test2.csv')
train = train.drop('Unnamed: 0',axis=1)
test = test.drop('Unnamed: 0',axis=1)
```

```python
train.corr()
```

## 모델 정의 및 학습

1. train에서 correlation 수치가 낮은 attribute를 삭제하고 학습을 진행해본다.
46.0972939909 -> 45.6896861661

```python
train_x = train.drop(['count','비','가시성','미세먼지10','미세먼지2.5'],axis = 1)
train_y = train['count']
```

```python
model=RandomForestRegressor(n_estimators=100)
model.fit(train_x,train_y)
```

```python
pred = model.predict(test.drop(['비','가시성','미세먼지10','미세먼지2.5'],axis=1))
submission = pd.read_csv(data_folder+'submission.csv')
submission['count'] = pred
submission.to_csv('/content/gdrive/MyDrive/data/BikeRental_Data/prediction3.csv',index = False)
```

2. 다양한 모델을 사용해본다

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
```

```python
train_x = train.drop(['count','비','가시성','미세먼지10','미세먼지2.5'],axis = 1)
train_y = train['count']
model_dt=DecisionTreeRegressor()
model_dt.fit(train_x,train_y)
model_knn = KNeighborsRegressor()
model_knn.fit(train_x,train_y)
```

```python
pred_dt = model_dt.predict(test.drop(['비','가시성','미세먼지10','미세먼지2.5'],axis=1))
submission = pd.read_csv(data_folder+'submission.csv')
submission['count'] = pred_dt
submission.to_csv('/content/gdrive/MyDrive/data/BikeRental_Data/prediction3_dt.csv',index = False)

pred_knn = model_knn.predict(test.drop(['비','가시성','미세먼지10','미세먼지2.5'],axis=1))
submission = pd.read_csv(data_folder+'submission.csv')
submission['count'] = pred_knn
submission.to_csv('/content/gdrive/MyDrive/data/BikeRental_Data/prediction3_knn.csv',index = False)
```
